```python
import os
import sys
sys.path.extend(['../../', '../../base'])
import re
import imageio
import cv2
import matplotlib.pyplot as plt
from base.utils.annotation_converter import AnnotationConverter
from base.utils.aida_annotation import AIDAnnotation
```

```python
wsi_path = '/Users/andreachatrian/Documents/Repositories/AIDA/dist/data/images/17_A047-4463_153D+-+2017-05-11+09.40.22.ndpi'
```

```python
converter = AnnotationConverter()
converter.by_overlap = False
aida_ann = AIDAnnotation('17_A047-4463_153D+-+2017-05-11+09.40.22.ndpi', 'test', ['epithelium', 'lumen', 'background'])
```

```python
gt_path = '/Volumes/A-CH-EXDISK/Projects/Dataset/train/17_A047-4463_153D+-+2017-05-11+09.40.22_TissueTrain_(1.00,30774,15012,3897,4556)/17_A047-4463_153D+-+2017-05-11+09.40.22_TissueTrain_(1.00,30774,15012,3897,4556)_mask.png'
gt = imageio.imread(gt_path)
plt.imshow(gt)
```

```python
# extract offset coords from area name
coords_pattern = '\((\w\.\w{1,3}),(\w{1,6}),(\w{1,6}),(\w{1,6}),(\w{1,6})\)'
coords_info = re.search(coords_pattern, os.path.basename(gt_path)).groups()  # tuple with all matched groups
print(coords_info)
area_x, area_y, area_w, area_h = tuple(int(num) for num in coords_info[1:])
```

```python
contours, labels = converter.mask_to_contour(gt, x_offset=area_x, y_offset=area_y)
```

```python
contoured_gt = gt.copy()
cv2.drawContours(contoured_gt, contours, -1, 50, 40)
plt.figure(figsize=(10,10))
plt.imshow(contoured_gt)
```

```python
for contour, label in zip(contours, labels):
    aida_ann.add_item(label, 'path')
    contour = contour.squeeze().astype(int).tolist()  # deal with extra dim at pos 1
    aida_ann.add_segments_to_last_item(contour)
```

```python
aida_ann_dir = '/Users/andreachatrian/Documents/Repositories/AIDA/dist/data/annotations'
aida_ann.dump_to_json(aida_ann_dir)
```
