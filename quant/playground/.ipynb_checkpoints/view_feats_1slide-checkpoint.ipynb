```python
import sys
sys.path.extend(['../', '../..'])
from pathlib import Path
import json
import copy
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA, FastICA
from sklearn.ensemble import IsolationForest
from sklearn.mixture import GaussianMixture
from quant.cluster import preprocess_features
```

```python
def fit_gaussians(x):
    gm = GaussianMixture(n_components=5)
    labels = gm.fit_predict(x)
    parameters = gm.get_params()
    return labels, parameters

def view_feats(feature_dir, feature_file):
    with open(feature_path, 'r') as features_file:
        x = pd.read_json(features_file, orient='split')
    xa = StandardScaler().fit_transform(x) # scikit-learn calls np.array on dataframe ?
    with_outliers = xa.shape[0]
    outlier_labels = IsolationForest().fit_predict(xa)  # outliers have label = -1
    xa = xa[np.where(outlier_labels == 1)]
    without_outliers = xa.shape[0]
    xa = PCA(n_components=2).fit_transform(xa)
    labels, parameters = fit_gaussians(xa)
    labels = np.array(labels)
    print(f"Number of outliers: {with_outliers - without_outliers}")
    colors = ['b', 'g', 'r', 'c', 'm']
    for i, label in enumerate(np.unique(labels)):
        ix = np.where(labels == label)
        plt.scatter(xa[ix, 0], xa[ix, 1], color=colors[i])
    return x, xa, params
```

```python
feature_dir = Path('/mnt/rescomp/projects/prostate-gland-phenotyping/WSI/data/features')
feature_path = feature_dir/'17_A047-4519_1614P+-+2017-05-11+09.50.49.json'
xa, params = view_feats(feature_dir, feature_path)
```
