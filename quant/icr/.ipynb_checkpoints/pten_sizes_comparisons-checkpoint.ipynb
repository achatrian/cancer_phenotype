# ICR size analysis

```python
from pathlib import Path
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from scipy import stats
import h5py
from tqdm import tqdm
```

```python
h_score_path = Path('/mnt/rescomp/projects/PTEN_TIFF/data/ICR_cytoplasmic_H_score.xlsx')
h_scores = pd.read_excel(h_score_path)
print(len(h_scores))
```

# list of PTEN cytoplasmic H-scores

```python
h_scores
```

```python
sizes_path = Path('/mnt/rescomp/projects/PTEN_TIFF/data/cell_size_2019-12-12.csv')
sizes = pd.read_csv(sizes_path)
sizes = sizes.loc[:, ~sizes.columns.str.contains('^Unnamed')]  # remove index column in excel
```

```python
sizes
```

```python
images = tuple(set(sizes['image']))
images, len(images) # missing: 602_-_PTEN_H&E_pyramid; V4032_-_A_-_PTEN_H&E_pyramid
```

```python
image_rows = []
for image in images:
    image_frame = sizes[sizes['image'] == image]
    image_rows.append(image_frame.mean(numeric_only=True))
image_sizes = pd.concat(image_rows, axis=1, keys=images).T
```

```python
# remove modifications to image names
image_sizes.index = list(slide_id.replace('_', ' ')[:-8] + '.czi' for slide_id in image_sizes.index)
```

```python
image_sizes
```

```python
# threshold h scores to assign each image to PTEN loss vs normals
group = []
for image, _ in image_sizes.iterrows():
    group.append('loss' if h_scores[h_scores['Case Id'] == image]['Cytoplasmic H score'].iloc[0] == 0 else 'normal')
image_sizes['pten'] = group
```

# statistics for mean difference: nuclear size

```python
t_nuclear, p_nuclear = stats.ttest_ind(
    image_sizes[image_sizes['pten'] == 'loss']['nuclear_size:mean'],
    image_sizes[image_sizes['pten'] == 'normal']['nuclear_size:mean'],
    equal_var=False
)
print(t_nuclear, p_nuclear)
```

```python
sns.catplot('pten', 'nuclear_size:mean', data=image_sizes, kind='box')
plt.title('nuclear size for pten loss vs normal')
plt.savefig('/home/andrea/Documents/Temp/Results/icr/nuclear_size.png')
```

# statistics for mean difference: cell size

```python
t_nuclear, p_nuclear = stats.ttest_ind(
    image_sizes[image_sizes['pten'] == 'loss']['cell_size:mean'],
    image_sizes[image_sizes['pten'] == 'normal']['cell_size:mean'],
    equal_var=False
)
print(t_nuclear, p_nuclear)
```

```python
sns.catplot('pten', 'cell_size:mean', data=image_sizes, kind='box')
plt.title('cell size for pten loss vs normal')
plt.savefig('/home/andrea/Documents/Temp/Results/icr/cell_size.png')
```

# comparison of nuclear areas for pten loss vs pten normal nuclei

```python
pten_loss_nucleus_areas, pten_normal_nucleus_areas = [], []
with h5py.File(Path('/mnt/rescomp/projects/PTEN_TIFF/data/cell_sizes.h5'), 'r') as areas_file:
    for area_id in tqdm(list(areas_file.keys())):
        roi_id, nucleus_box = area_id.split(':')  # roi_id:x_y_w_h
        # TODO need to adapt roi_id so it's compatible with keys in h5 file
        roi_id = roi_id.replace('_', ' ')[:-8] + '.czi' # TODO test
        if h_scores[h_scores['Case Id'] == roi_id]['Cytoplasmic H score'].iloc[0] == 0:
            pten_loss_nucleus_areas.extend(areas_file[area_id])
        else:
            pten_normal_nucleus_areas.extend(areas_file[area_id])
```

```python
n_bins = 100
range_ = (0, 1000)
plt.hist(pten_loss_nucleus_areas, bins=n_bins, range=range_, cumulative=False, density=True, histtype='stepfilled', alpha=0.5, label='PTEN loss')
plt.hist(pten_normal_nucleus_areas, bins=n_bins, range=range_, cumulative=False, density=True, histtype='stepfilled', alpha=0.4, label='PTEN normal')
plt.legend(loc='upper left')
plt.ylabel('density')
plt.xlabel('nucles size (pixels)')
plt.title('nuclear area distributions - PTEN loss vs normal')
plt.savefig('/home/andrea/Documents/Temp/Results/icr/nuclear_size_hists.png')
```

```python
len(pten_loss_nucleus_areas), len(pten_normal_nucleus_areas)
```

```python
# statistics for mean difference: cell size
t_nuclear, p_nuclear = stats.ttest_ind(
    pten_loss_nucleus_areas,
    pten_normal_nucleus_areas,
    equal_var=False
)
print(t_nuclear, p_nuclear)
```
